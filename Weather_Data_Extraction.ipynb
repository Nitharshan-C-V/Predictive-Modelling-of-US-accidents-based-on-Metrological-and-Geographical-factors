```python
T2M=[]
```

```python
for i in range(len(Lat_T)):
  URL_T="https://power.larc.nasa.gov/api/temporal/daily/point?start="+str(Date_T[i])+"&end="+str(Date_T[i])+"&latitude="+str(Lat_T[i])+"&longitude="+str(Long_T[i])+"&community=re&parameters=T2M&time-standard=lst"
  response = requests.get(URL_T)
  T2M.append(response.json()['properties']['parameter']['T2M'][Date_T[i]])
```

```python
len(T2M)
```

```python
T2M[0:5]
```

```python
T2M[-5:]
```

Need to Convert C to F

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

for i in range(len(T2M)):
  T2M[i] = celsius_to_fahrenheit(T2M[i])
```

```python
T2M[0:5]
```

```python
#Rounding OFF to 1 decimal place
for i in range(len(T2M)):
  T2M[i]=round(T2M[i],1)
```

```python
T2M[0:5]
```

```python
T2M[-5:]
```

```python
#T2M has Final Temperature Missing Values
Final_M_TEMP=T2M.copy()
```

```python
len(Final_M_TEMP)
```

```python
Temp_DIC={"Temperature(F)" : Final_M_TEMP}
```

```python
Temp_df_final=Temp_df.drop(columns=['Temperature(F)'])
```

```python
Temp_df_final['Temperature(F)'] = Temp_DIC['Temperature(F)']
```

```python
Temp_df_final.info()
```

```python
Temp_df_final
```

```python
Temp_df_final.to_csv("Temp_df_final",index=False)
```
